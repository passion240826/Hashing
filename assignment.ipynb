# 2번

### 2-1

테이블의 크기는 1204이지만, 해시함수로의 결과 범위는 0 ~ 1023 이므로, 1024 ~ 1203 위치는 사용되지 않음
따라서
- 테이블 공간의 낭비 발생으로 비효율적임 

### 2-2

h(x) = x*random(0,10)%m 함수는 동일한 키값에 대해 동일한 해시값이 나오지 않기 때문에 h(x)는 해시 함수가 될 수 없다.
- 같은 키값에 대해 해시값이 변함
- 저장위치와 탐색 위치가 달라짐
- 검색이 불가능해짐
